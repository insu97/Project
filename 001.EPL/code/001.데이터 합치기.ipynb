```python
import pandas as pd
```

```python
df = pd.DataFrame()

a = 2000
while(a != 2024):
    if a == 2000 :
        df = pd.read_csv("../data/Season %s-%s.csv" % (a, a+1))
        print(a, len(df), len(df)/380)
        a += 1
    else :
        df = pd.concat([df,pd.read_csv("../data/Season %s-%s.csv" % (a, a+1), encoding='cp949')], ignore_index=True)
        print(a, len(df), len(df)/380)
        a += 1
```
